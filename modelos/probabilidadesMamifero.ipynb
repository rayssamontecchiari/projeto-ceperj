```python
import pandas as pd

# seleciona os dados do arquivos
# data frame (df) - molde de dados
dados = pd.read_csv('CSV_Mamíferos_Cortada.csv')
dados.head()
```

```python
Y_df = dados['mamifero mais atropelado']
X_df = dados[["Mes", "Estacao", 
              "Trecho", "Sentido", "Trecho macro",
              "Tipo De Pistas", "Numero de Pistas",
              "Tipo de Pavimento", "Numero de Faixas",
              "Velocidade Maxima", "Tipo da Chuva","Intervencao",
              "Vazamento", "Agua?", "Vegetacao Baixa?","capimAlto",
              "Arbustos","Floresta","Local","Com filhotes?",
              "Faixa Horario da Coleta","Faixa de Quilometragem"]]

features = ["Mes", "Estacao", 
              "Trecho", "Sentido", "Trecho macro",
              "Tipo De Pistas", "Numero de Pistas",
              "Tipo de Pavimento", "Numero de Faixas",
              "Velocidade Maxima", "Tipo da Chuva","Intervencao",
              "Vazamento", "Agua?", "Vegetacao Baixa?","capimAlto",
              "Arbustos","Floresta","Local","Com filhotes?",
              "Faixa Horario da Coleta","Faixa de Quilometragem"]

count0 = 0
count1 = 0

for i in Y_df:
    if i==0:
        count0 = count0 + 1
        
    if i==1:
        count1 = count1 + 1
    
    
print(count0, "x", count1)

valores = []

for i in X_df:
    auxi = X_df[i].unique()
    valores.append(auxi)
    
print("Probabilidade de obtermos 1:", count1/len(Y_df))
print("Probabilidade de obtermos 0:", count0/len(Y_df))
```

Quantas vezes cada uma das características aparecem quando a classe é sim, dividido pelo total de vezes que obtemos a classes sim

```python
def calculaProb(nomeAtributo, posicaoValores):
    prob = []
    
    for i in valores[posicaoValores]:
        auxi1 = 0
        auxi0 = 0
        
        for j in dados.index: # para cada linha do dataset
            if dados[nomeAtributo][j] == i and dados["mamifero mais atropelado"][j] == 1:
                auxi1 = auxi1 + 1
            if dados[nomeAtributo][j] == i and dados["mamifero mais atropelado"][j] == 0:
                auxi0 = auxi0 + 1   
        
        prob1 = auxi1/count1 
        prob0 = auxi0/count0
        prob.append({nomeAtributo: i, "prob ser 1":prob1, "prob ser 0": prob0})
    
    return(prob)
        
```

```python
def transformaDados(atri, array):
    nome = []
    prob1 = []
    prob0 = []
    
    for i in array:
        nome.append(i[atri])
        prob1.append(i["prob ser 1"])
        prob0.append(i["prob ser 0"])

    probDF = pd.DataFrame()
    probDF[atri] = nome
    probDF["prob ser 1"] = prob1
    probDF["prob ser 0"] = prob0
    
    return probDF
    
```

```python
for i in range(len(features)):
    # print(valores[i])
    # print(features[i])
    prob = calculaProb(features[i], i)
    probDf = transformaDados(features[i], prob)
    print(probDf, "\n")
```
